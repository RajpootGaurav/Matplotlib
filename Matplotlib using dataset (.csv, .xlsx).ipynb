```python
# Passing ist data to plot
x = [0, 1, 2, 3, 4, 5]
y = [0, 10, 20, 30, 40, 50]

plt.plot(x, y)
plt.show()
```

```python
# plotting data from files read using pandas

# Gross World Product Data since 1970 - colelcted from wikipedia
data = pd.read_csv("gwp.csv")

data
```

```python
# Plotting pandas dataframe by passing dataframes columns as x and y axis

plt.plot(data['Year'], data['GWP'])
plt.show()
```

```python
# Data Visualization by calling plot function directly on the dataframe

# Set Year as index
data1 = data.set_index("Year")
# Display new dataframe
data1
```

```python
# Creating a plot by calling plot directly on the dataframe

data1['GWP'].plot()
plt.show()
```

# Forth part

```python
# set matpltolob to draw plots inline
%matplotlib inline

# import matplotlib pyplot
import matplotlib.pyplot as plt

# import numpy
import numpy as np 

# import pandas
import pandas as pd
```

```python
# Online Retail Data Set
# *from https://archive.ics.uci.edu/ml/datasets/online+retail

# import dataset (excel file)
data = pd.read_excel("online-retail.xlsx")
```

```python
# Explore Data
data.head()
```

```python
sales_by_product = pd.DataFrame({'total_sales' : data.groupby( ["StockCode"] )['Quantity'].
                             sum()}).reset_index().set_index(['StockCode'])
sales_by_product.head()
```

```python
plt.figure(figsize=(12,5));
chart = sales_by_product.total_sales.plot(kind='bar')
```

```python
top10 = sales_by_product.sort_values("total_sales")[:10]

plt.figure(figsize=(12,5));
chart = top10.total_sales.plot(kind='bar')
```

```python
top10 = sales_by_product.sort_values("total_sales", ascending=False)[:10]

plt.figure(figsize=(12,5));
chart = top10.total_sales.plot(kind='bar');
```
